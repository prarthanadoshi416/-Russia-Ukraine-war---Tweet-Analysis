## Hastags

**re** i.e. Regular Expression is a sequence of characters that forms a search pattern. RegEx can be used to check if a string contains the specified search pattern.

```python
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
text = data['content'][100]
text
```

The findall() function returns a list containing all matches.

r to read the particular string

\w	Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)

\w+ means all the words

```python
ht = re.findall(r"#(\w+)", text)
ht
```

```python
def hashtag_extract(text_corpus):
  hashtag=[] # make hashtag as an empty list
  for text in text_corpus:
    ht=re.findall(r"#(\w+)",text)
    hashtag.append(ht)  #. append() method takes an object as an argument and adds it to the end of an existing list. 
  return hashtag
```

**Function to count the most frequent Hashtags**

```python
def hashtag_freq(hashtag):
  a=nltk.FreqDist(hashtag)
  #The FreqDist function gives the user the frequency distribution of all the words in the text. 
  df = {'Hashtag':list(a.keys()),'Freq': list(a.values())}
  d=pd.DataFrame(df)   # converting dictionary to dataframe
  d=d.nlargest(columns="Freq",n=30) #sorthing the dataframe. Return the first n rows with the largest values in columns, 
                                    #in descending order. The columns that are not specified are returned as well, but not used for ordering.
                                    #n is the number or rows ordered
  return d
```

```python
hashtags=hashtag_extract(data['content'])
```

```python
# hashtags
```

```python
# Since the hashtags are list of list so we have to make it to single list. Sum of all Hashtags in an empty list
hashtags = sum(hashtags,[])
# hashtags
```

```python
hash = hashtag_freq(hashtags)
hash
```

```python
plt.figure(figsize=(36,17)) #using matlplot lib
```

```python
ax = sns.barplot(data = hash, x="Hashtag", y="Freq")
plt.xticks(rotation=90)
plt.show()
```

```python
import string
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords") 
from nltk.stem import PorterStemmer
```

```python
text = data['content'][100]
text
```

```python
# All numbers, puctuations and links has to be removed. 
# 1. First we will convert all text to small alphabets
text = str(text).lower() 
# 2. Now we will remove the url from the tweet. re.sub is regular expression substitution
# \S+ mean it includes all the capital/small letter and special characters without space, the string DOES NOT contain a white space character
text = re.sub('https://\S+|www\.\S+','',text) #there should be no space before and after or(|)
# 3. Now we will remove all the punctuations. For that we have to import the string library which has a list of punctuations
text = re.sub(r'[^\w\s]','',text) #\w\s looks for words or whitespaces. 
                                  # We can select the opposite of this (i.e., anything that isn’t a word or whitespace) using the ^ character.
#4. Now we will remove all the \n and \t and all
text = re.sub('\n|\t','',text) 
#5. Now we will remove all the digits
text = re.sub('\d','',text) 
text
```

```python
#6. Now we will remove all the stopwords.  A stop word is a commonly used word (such as “the”, “a”, “an”, “in”) that a search engine has been programmed to ignore, 
# both when indexing entries for searching. For that we import and download stopwords
from nltk.corpus import stopwords
nltk.download("stopwords") #run only first time to install stopwords. Once installed comment it.
stop_words = stopwords.words('english')
print(stop_words)
len(stop_words)
```

```python
# Now we will define a function for cleaning of text for all tweets
def clean(text):
  text = str(text).lower() 
  text = re.sub('https://\S+|www\.\S+','',text) 
  text = re.sub(r'[^\w\s]','',text) 
  text = re.sub('\n|\t','',text) 
  text = re.sub('\d','',text) 
  stop_words = stopwords.words('english')
  stop_words = set(stop_words)
  len(stop_words)
  text = [word for word in text.split(' ') if word not in stop_words]
  text = " ".join(text)
  ps = PorterStemmer()
  text=[ps.stem(word) for word in text.split(' ')]
  text = " ".join(text)
  return text
```

```python
data['clear_tweet']=data['content'].apply(clean)
```

```python
text = data['clear_tweet'][10]
text
```

```python
analysis = TextBlob(text)
analysis.sentiment.polarity
```

```python
def sen_analysis(text):
  analysis = TextBlob(text)
  if(analysis.sentiment.polarity > 0):
    return 1
  elif(analysis.sentiment.polarity == 0):
    return 0
  else:
    return -1
```

```python
# Making a column name sentiment and is applies to clear_tweets
data['sentiment'] = data['clear_tweet'].apply(sen_analysis)

# # Making a column name sentiment and is applies to clear_tweets
# data['sentiment'] = data['clear_tweet'].apply(lambda x:sen_analysis(x))
```
